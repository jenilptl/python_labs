## Practical 2: Inventory Management

### Problem Definition
Remove out-of-stock items and calculate total inventory value.

### Sample Input
```python
inventory = {
    'pen': (10, 50),
    'book': (100, 0),
    'pencil': (5, 100)
}
```

### Sample Output
```
{'pen': (10, 50), 'pencil': (5, 100)}
Total Value: 1000
```

````python
inventory = {'pen': (10, 50), 'book': (100, 0), 'pencil': (5, 100)}
filteredinventory = {item: (price, qty) for item, (price, qty) in inventory.items() if qty > 0}
totalvalue = sum((price * qty for price, qty in filteredinventory.values()))
print(filteredinventory)
print(f'Total Value: {totalvalue}')
````

## Practical 3: Email Domain Extractor

### Problem Definition
Extract email domains and count occurrences.

### Sample Input
```python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
```

### Sample Output
```
{'gmail.com': 2, 'yahoo.com': 1}
```

````python
emails = ['a@gmail.com', 'b@yahoo.com', 'c@gmail.com']
domcnt = {}
for email in emails:
    domain = email.split('@')[1]
    domcnt[domain] = domcnt.get(domain, 0) + 1
print(domcnt)
````

## Practical 4: Shopping Cart Billing

### Problem Definition
Calculate item-wise cost and total bill.

### Sample Input
```
cart = [('apple',2), ('banana',3)]
prices = {'apple':50, 'banana':10}
```

### Sample Output
```
{'apple': 100, 'banana': 30}
Total Bill: 130
```

````python
cart = [('apple', 2), ('banana', 3)]
prices = {'apple': 50, 'banana': 10}
dic = {}
sum = 0
for i in cart:
    item = i[0]
    price = prices[item]
    dic[item] = price * i[1]
    sum = sum + price * i[1]
print(sum)
print(dic)
````

## Practical 5: Character Position Index

### Problem Definition
Store positions of each character in a string.

### Sample Input
```python
s = "banana"
```

### Sample Output
```
{'b': [0], 'a': [1, 3, 5], 'n': [2, 4]}
```

````python
s = 'banana'
result = {}
for i, ch in enumerate(s):
    if ch not in result:
        result[ch] = []
    result[ch].append(i)
print(result)
````

## Practical 6: Survey Data Aggregation

### Problem Definition
Count responses city-wise and choice-wise.

### Sample Input
```python
responses = [
    ('Amit','Rajkot','Yes'),
    ('Neha','Rajkot','No'),
    ('Ravi','Ahmedabad','Yes')
]
```

### Sample Output
```
{'Rajkot': {'Yes': 1, 'No': 1}, 'Ahmedabad': {'Yes': 1}}
```

````python
responses = [('Amit', 'Rajkot', 'Yes'), ('Neha', 'Rajkot', 'No'), ('Ravi', 'Ahmedabad', 'Yes')]
result = {}
for , city, response in responses:
    if city not in result:
        result[city] = {}
    if response not in result[city]:
        result[city][response] = 0
    result[city][response] += 1
print(result)
````
