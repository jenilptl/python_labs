### 2) Reverse Each Word
- Input: `Python is Easy` → `nohtyP si ysaE`
- Input: `Hello World` → `olleH dlroW`
- Input: `Learn Python Fast` → `nraeL nohtyP tsaF`

```python
def reversewords(sentence):
    words = sentence.split()
    reversedwords = [word[::-1] for word in words]
    return ' '.join(reversedwords)
print(reversewords('Python is Easy'))
print(reversewords('Hello World'))
print(reversewords('Learn Python Fast'))
```

### 10) Toggle Case (without using str.swapcase())
- `PyThOn` → `pYtHoN`
- `HELlo` → `helLO`
- `Python` → `pYTHON`

```python
string = 'python IS A PROGRAMMING language'
newstring = []
for i in string:
    if i.isupper():
        i.lower()
        newstring.append(i)
    else:
        i.upper()
        newstring.append(i)
print(''.join(newstring))
```
