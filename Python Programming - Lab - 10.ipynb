### 08) WAP to create an UDF named frequency to count occurances of the specific word in a given text file.

```python
def freq(file, word):
    with open(file, 'r') as df:
        occ = df.read().count(word)
        print(occ)
file = 'demo.txt'
word = 'ggg'
freq(file, word)
```

### 09) WAP to get the score of five subjects from the user, store them in a file. Fetch those marks and find the highest score.

```python
with open('marks.txt', 'w') as f:
    for i in range(1, 6):
        mark = int(input(f'Enter marks for subject {i}: '))
        f.write(str(mark) + '\n')
with open('marks.txt', 'r') as f:
    marks = [int(line.strip()) for line in f]
highest = max(marks)
print('Highest score is:', highest)
```

### 10) WAP to write first 100 prime numbers to a file named primenumbers.txt 
(Note: each number should be in new line)

```python
def isprime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True
with open('primenumbers.txt', 'w') as file:
    count = 0
    num = 2
    while count < 100:
        if isprime(num):
            file.write(str(num) + '\n')
            count += 1
        num += 1
```

### 12) WAP to replace word1 by word2 of a text file. Write the updated data to new file.

```python
def change(word1, word2):
    with open('demo.txt', 'r') as fd:
        data = fd.read()
        rpdata = data.replace(word1, word2)
        rpdata = data.replace(word2, word1)
    with open('nwdemo.txt', 'w') as kg:
        kg.write(rpdata)
wd1 = 'helo'
wd2 = 'world'
change(wd1, wd2)
```
