### 12) WAP to find whether the given number is Prime or not.

```python
n = int(input('Enter a number: '))
flag = 1
for i in range(2, int(n / 2 + 1)):
    if n % i == 0:
        flag = 0
if flag == 1:
    print(n, 'is an prime number')
else:
    print(n, 'is NOT prime number')
```

### 13) WAP to print sum of digits of given number.

```python
n = int(input('Enter a number: '))
armsum = 0
temp = n
while temp != 0:
    digit = temp % 10
    armsum = armsum + digit
    temp //= 10
print('sum of digits', armsum)
```

### 14) WAP to check whether the given number is Palindrome or not.

```python
n = int(input('Enter a number: '))
armsum = 0
temp = n
while temp != 0:
    digit = temp % 10
    armsum = armsum * 10 + digit
    temp //= 10
if armsum == n:
    print(n, 'is an palindrome number')
else:
    print(n, 'is NOT palindrome number')
```

### 15) WAP to check whether the given number is an Armstrong Number or not.

```python
n = int(input('Enter a number: '))
armsum = 0
temp = n
count = 0
while temp != 0:
    count += 1
    temp //= 10
temp = n
while temp != 0:
    digit = temp % 10
    armsum += digit ** count
    temp //= 10
if armsum == n:
    print(n, 'is an Armstrong number')
else:
    print(n, 'is NOT armstrong num')
```

### 16) WAP to print all the perfect numbers between 1 to n.

```python
n = int(input('Enter n: '))
for num in range(1, n + 1):
    sum = 0
    for i in range(1, num):
        if num % i == 0:
            sum += i
    if sum == num:
        print(num)
```
