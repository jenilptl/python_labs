### 05) Create a bank account class with methods to deposit, withdraw, and check balance.

```python
class BankAccount:
    def __init__(self, account_number, initial_balance=0):
        self.account_number = account_number
        self.balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount:.2f}. New balance: ${self.balance:.2f}")
        else:
            print("Invalid deposit")

    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                print(f"Withdrew ${amount:.2f}. New balance: ${self.balance:.2f}")
            else:
                print("Insufficient balance.")
        else:
            print("Invalid withdrawal")

    def check_balance(self):
        print(f"Current balance {self.account_number}: ${self.balance:.2f}")   
```

### 06) Create a class for managing inventory that includes attributes such as item name, price, quantity, and methods to add, remove, and update items.

```python
class Inventory:
    def __init__(self):
        self.items = {}

    def add_item(self, item_name, price, quantity):
        if item_name in self.items:
            self.items[item_name]['quantity'] += quantity
            print(f"Updated {item_name}: {quantity} added. New total: {self.items[item_name]['quantity']}")
        else:
            self.items[item_name] = {'price': price, 'quantity': quantity}
            print(f"Added {item_name} with quantity {quantity} at ${price:.2f} each.")

    def remove_item(self, item_name, quantity):
        if item_name not in self.items:
            print(f"Error: {item_name} not found in inventory.")
            return
        if self.items[item_name]['quantity'] < quantity:
            print(f"Error: Cannot remove {quantity} {item_name}(s). Only {self.items[item_name]['quantity']} available.")
        else:
            self.items[item_name]['quantity'] -= quantity
            if self.items[item_name]['quantity'] == 0:
                del self.items[item_name]
                print(f"{item_name} removed from inventory.")
            else:
                print(f"Removed {quantity} {item_name}(s). Remaining: {self.items[item_name]['quantity']}")

    def update_item(self, item_name, price=None, quantity=None):
        if item_name not in self.items:
            print(f"Error: {item_name} not found in inventory.")
            return
        if price is not None:
            self.items[item_name]['price'] = price
        if quantity is not None:
            self.items[item_name]['quantity'] = quantity
        print(f"{item_name} updated: Price = ${price:.2f}, Quantity = {quantity}")

    def display_inventory(self):
        if not self.items:
            print("Inventory is empty.")
            return
        print("\nCurrent Inventory:")
        print(f"{'Item':<15} {'Price':<10} {'Quantity':<10}")
        print("-" * 35)
        for name, details in self.items.items():
            print(f"{name:<15} ${details['price']:<9.2f} {details['quantity']:<10}")
```

### 09) Define Time class with hour and minute as data member. Also define addition method to add two time objects.

```python
class Time:
    def __init__(self, hour=0, minute=0):
        self.hour = hour
        self.minute = minute

    def add(self, other):
        total_minutes = self.minute + other.minute
        extra_hour = total_minutes // 60
        minutes = total_minutes % 60
        hours = self.hour + other.hour + extra_hour
        return Time(hours, minutes)

    def __str__(self):
        return f"{self.hour:02d}:{self.minute:02d}"

t1 = Time(5, 40)
t2 = Time(3, 30)
result = t1.add(t2)
print(result)
```
