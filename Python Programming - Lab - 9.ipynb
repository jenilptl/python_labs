### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def isprime(n):
    if n <= 1:
        return 0
    if n <= 3:
        return 1
    if n % 2 == 0 or n % 3 == 0:
        return 0
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return 0
        i += 6
    return 1
k = isprime(11)
k
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def primlist(n1, n2):
    plist = []
    t = 0
    for i in range(n1, n2):
        for j in range(2, i):
            if i % j == 0:
                break
        else:
            plist.append(i)
    return plist
pl = primlist(10, 20)
pl
```

### 05) Write a function that returns True if the given string is Palindrome or False otherwise.

```python
def pallist(str):
    tlist = str[::-1]
    if tlist == str:
        return True
    return False
pl = pallist('abbdbba')
pl
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
def sumfirstelements(tupleslist):
    total = 0
    for t in tupleslist:
        total += t[0]
    return total
n = int(input('Enter number of tuples: '))
tupleslist = []
for i in range(n):
    t = tuple(map(int, input(f'Enter tuple {i + 1}: ').split()))
    tupleslist.append(t)
result = sumfirstelements(tupleslist)
print('Sum of first elements:', result)
```

### 08) Write a recursive function to find nth term of Fibonacci Series.

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)
term = fibonacci(5)
term
```

### 10) Write a function to get the sum of the scores ending with zero.
#### Example : scores = [200, 456, 300, 100, 234, 678] 
#### Ans = 200 + 300 + 100 = 600

```python
def sumdiv(numbers):
    total = 0
    for n in numbers:
        if n % 10 == 0:
            total += n
    return total
sum = sumdiv([1, 10, 9, 20])
sum
```

### 11) Write a function to invert a given Dictionary.
#### hint: keys to values & values to keys
#### Before : {'a': 10, 'b':20, 'c':30, 'd':40}
#### After : {10:'a', 20:'b', 30:'c', 40:'d'}

```python
def invert(dic):
    invdi = {}
    for k, v in dic.items():
        invdi[v] = k
    return invdi
indic = invert({'a': 10, 'b': 20, 'c': 30, 'd': 40})
indic
```

### 12) Write a function that returns the number of uppercase and lowercase letters in the given string. 
#### example : Input : s1 = AbcDEfgh ,Ouptput : no_upper = 3, no_lower = 5

```python
def letter(s):
    lw, up = (0, 0)
    for ch in s:
        if ch.islower():
            lw = lw + 1
        else:
            up += 1
    return (up, lw)
up, lw = letter('JenPtl')
(up, lw)
```

### 16) Write udfs to call the functions with following types of arguments:
1. Positional Arguments
2. Keyword Arguments
3. Default Arguments
4. Variable Legngth Positional(*args) & variable length Keyword Arguments (**kwargs)
5. Keyword-Only & Positional Only Arguments

```python
def positional(a, b, c):
    print(a, b, c)
positional(1, 2, 3)

def keyword(a, b, c):
    print(a, b, c)
keyword(a=1, c=3, b=2)

def default(a, b, c=0):
    print(a, b, c)
default(1, 2)

def var(a, b, *var):
    print(a, b, var)
var(1, 2, 3, 4, 5, 6)

def var2(a, b, **kwargs):
    print(a, b, kwargs)
var2(1, 2, c=3, d=4, e=5)

def posi(a, b, c, /):
    print(a, b, c)
posi(1, 2, 3)

def keywordonly(*, a, b, c):
    print(a, b, c)
keywordonly(a=1, c=3, b=2)
```
