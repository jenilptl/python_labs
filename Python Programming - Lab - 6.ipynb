### 01) WAP to find sum of all the elements in a List.

```python
li = input('enter a list')
li = list(li)
sum = 0
for i in li:
    sum = sum + int(i)
print(sum)
```

### 02) WAP to find largest element in a List.

```python
li = [1, 3, 5, 7, 9]
sum = li[0]
for i in li:
    if sum < i:
        sum = i
print(sum)
```

### 12) WAP to count occurrences of an element in a list.

```python
li = list(input('Enter list elements: '))
ele = input('Enter element to count: ')
count = 0
for x in li:
    if x == ele:
        count += 1
print('Occurrences:', count)
```

### 14) WAP to extract elements with frequency greater than K.

```python
li = list(map(int, input('Enter list: ').split()))
k = int(input('Enter K: '))
res = []
for i in li:
    if li.count(i) > k and i not in res:
        res.append(i)
print(res)
```

### 15) WAP to create a list of squared numbers from 0 to 9 with and without using List Comprehension.

```python
li = (i ** 2 for i in range(0, 10))
li = list(li)
l2 = []
for i in range(0, 10):
    l2.append(i ** 2)
print(li)
print(l2)
```

### 16) WAP to create a new list (fruit whose name starts with 'b') from the list of fruits given by user.

```python
li = list(map(str, input('enter a elemnts:').split(',')))
ans = []
for i in li:
    if i[0] == 'b':
        ans.append(i)
print(ans)
```

### 17) WAP to create a list of common elements from given two lists.

```python
li = list(map(int, input('enter a elemnts:').split(',')))
ck = list(map(int, input('enter a elemnts:').split(',')))
cm = []
for i in li:
    for j in ck:
        if i == j:
            cm.append(i)
cm = list(set(cm))
print(cm)
```
