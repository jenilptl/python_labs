### 04) WAP to plot a line chart showing temperature variation during a week. Mark the lowest and highest temperatue.

```python
Days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
Temperature = [20, 32, 31, 33, 34, 43, 36]
```

```python
import matplotlib.pyplot as plt 

min_temp = min(Temperature)
max_temp = max(Temperature)

min_day = Days[Temperature.index(min_temp)]
max_day = Days[Temperature.index(max_temp)]

plt.plot(Days, Temperature)

plt.scatter(min_day, min_temp, color='green', s=100, label='Lowest Temp')
plt.scatter(max_day, max_temp, color='red', s=100, label='Highest Temp')

plt.show()
```

### 06) WAP to create bar chart showing the marks obtained by students in different subjects.
### The program should also display the marks on the top of the respective bar.

```python
Subjects = ["Python", "DBMS", "SE", "PS", "PE-1"]
Marks = [78, 85, 72, 80, 90]
```

```python

bars = plt.bar(Subjects, Marks, color='skyblue')

for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2, 
        height,                          
        str(height),                    
        ha='center', va='bottom'
    )

plt.show()
```

### 07) WAP to create a horizontal bar chart representing the number of books sold in different categories. 
### The program should also display the number of books sold at the end of the respective bar.

```python
Categories = ["Fiction", "Science", "History", "Technology", "Biography"]
Sold = [120, 90, 70, 110, 60]
```

```python
bars = plt.barh(Categories, Sold)

for bar in bars:
    width = bar.get_width()
    plt.text(
        width,                    
        bar.get_y() + bar.get_height()/2, 
        str(width),
        va='center', ha='left'
    )

plt.show()
```
