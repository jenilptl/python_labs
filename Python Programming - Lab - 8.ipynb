### 06) WAP to Sort a dictionary in ascending order of its values.
Hint: in sorted(), use udf for key parameter


**Sample Input:**
```
{'Math': 70, 'Physics': 85, 'Chemistry': 60}

```
**Sample Output:**
```
{'Chemistry': 60, 'Math': 70, 'Physics': 85}
```

````python
n = int(input('enter a num'))
a = {}
for i in range(n):
    key = input('enter num')
    value = int(input('enter a num'))
    a[key] = value
p = a.keys()
k = sorted(a.values())
ans = {}
for j in p:
    for i in k:
        ans[j] = i
ans
````

### 07) WAP to find the keys having maximum and minimum values in a dictionary.

**Sample Input:**
```
{'A': 50, 'B': 90, 'C': 30}
```
**Sample Output:**
```
Max Value Key: B
Min Value Key: C

```

````python
data = {'A': 50, 'B': 90, 'C': 30}
maxkey = max(data, key=data.get)
minkey = min(data, key=data.get)
print('Max Value Key:', maxkey)
print('Min Value Key:', minkey)
````
